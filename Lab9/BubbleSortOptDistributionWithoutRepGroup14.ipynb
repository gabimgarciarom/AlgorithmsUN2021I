# Bubble sort not optimized time complexity histograms for permutations of $n$ different elements 

## Best case - increasingly sorted array ($\frac{n^2-n}{2}$ comparisons + $0$ swaps) - $t_b(n) = \frac{n^2-n}{2} = \Theta(n^2) $

## Worst case - decreasingly sorted array ($\frac{n^2-n}{2}$ comparisons + $\frac{n^2-n}{2}$  swaps) - $t_w(n) = n^2-n = \Theta(n^2) $

## Average case - expected number of comparisons + swaps on random unifrom permutation of $n$ different elements $t_a(n) = \frac{3}{4}(n^2-n)= \Theta(n^2) $



##Function Bubblesort##
Definiton of the bubblesort optimized is the following:

```python
def bubbleSort(arr):
    steps = 0
    n = len(arr)
 
    # Traverse through all array elements
    for i in range(n):
 
        # Last i elements are already in place
        for j in range(0, n-i-1):
            changes = 0
            # traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            steps +=1
            if arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j]
                changes +=1
                steps +=1
        if changes == 0:
          break
    return(steps)
 
```

We define a function to create the histograms of the time distributions and it's analysis:

```python
import matplotlib.pyplot as plt
import numpy as np 
import math
def n_analysis (n):
  perm = permutations(range(1,n+1))

  times =  [] 
  # Print the obtained permutations
  for i in list(perm):
      i = list(i)
      times.append(bubbleSort(i))

  b = np.array(range(min(times)-1,max(times)+2) ) +0.5
  plt.hist(times,bins=b)
  plt.title("Bubble Sort")
  plt.xlabel("Steps =comparisons + swaps ")
  plt.ylabel("Frequency")
  plt.show()
      
  print("n",n)  
  print("number of permutations n! =",math.factorial(n)) 
  print("exp min =",min(times))
  print("min formula (n-1)*n/2 =",(n-1)*n/2)
  print("exp max =",max(times))
  print("max formula (n-1)*n =",(n-1)*n)
  print("exp mean =",np.mean(times))
  print("mean formula for non optimized algorithm 3/4*n**2+17/4*n-4 =",(3/4)*(n**2-n))
  print("stdev",np.std(times))
```


Average case - expected number of comparisons + swaps on random unifrom permtation of $n$ diffrent elements $t_a(n) = \frac{3}{4}(n^2-n)= \Theta(n^2) $

```python
# Average case on random unifrom permtation of  n 3/4 (**2 -n) 
import pandas as pd
import scipy.linalg as la

n = np.array([3.0,4.0,5.0,6.0])    # Number n of sorted elments 
t = np.array([25/6,95/12,191/15,335/18]) # Average comparisons + swaps for the corresponding n 

data = pd.DataFrame({'n': n, 't': t})
data

```

```python
# Interpolating a polynomial of degree 3

X = np.column_stack([n**0,n**1,n**2,n**3])
a = la.solve(X,t)
a
```

We define a function to create the histograms of the time distributions and it's analysis:

```python
def n_repeat_analysis(n) :
  perm = product(range(1,n+1), repeat = n)

  times =  [] 
  # Print the obtained permutations
  for i in list(perm):
      i = list(i)
      times.append(bubbleSort(i))

  b = np.array(range(min(times)-1,max(times)+2) ) +0.5
  plt.hist(times,bins=b)
  plt.title("Bubble Sort")
  plt.xlabel("Steps = comparisons + swaps ")
  plt.ylabel("Frequency")
  plt.show()
      
  print("n",n)  
  print("number of permutations ,n**n = ",n**n)
  print("exp min =",min(times))
  print("min formula (n-1)*n/2 =",(n-1)*n/2)
  print("exp max =",max(times))
  print("max formula (n-1)*n =",(n-1)*n)
  print("exp mean =",np.mean(times))
  print("mean formula for non optimize algorithm (3/4)*(n**2)-n+(1/4) =",(3/4)*(n**2)-n+(1/4))
  print("stdev",np.std(times))
```
